```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("USD_BRL_hist.csv")
df["Data"] = pd.to_datetime(df["Data"], dayfirst=True)

plt.figure(figsize=(12,6))
plt.plot(df["Data"], df["USD_BRL"], color="darkred", linewidth=2.5, linestyle="-")
plt.title("Cotação do Dólar em Reais ao longo do tempo", fontsize=18, fontweight="bold", color="navy")
plt.xlabel("Data", fontsize=14, color="black")
plt.ylabel("Valor em BRL", fontsize=14, color="black")
plt.grid(True, linestyle="--", alpha=0.6)

plt.savefig("grafico_linha.png", dpi=300, bbox_inches="tight") 
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("USD_BRL_hist.csv")
df["Data"] = pd.to_datetime(df["Data"], dayfirst=True)

df["Ano"] = df["Data"].dt.year
media_ano = df.groupby("Ano")["USD_BRL"].mean()

plt.figure(figsize=(10,6))
media_ano.plot(kind="bar", color="seagreen", edgecolor="black")
plt.title("Média anual da cotação do dólar", fontsize=18, fontweight="bold", color="darkgreen")
plt.xlabel("Ano", fontsize=14, color="black")
plt.ylabel("Valor médio (BRL)", fontsize=14, color="black")
plt.xticks(rotation=45, fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.savefig("grafico_barra.png", dpi=300, bbox_inches="tight")  
plt.show()


```

```python
import pandas as pd
import plotly.graph_objects as go

df = pd.read_csv("USD_BRL_hist.csv")
df["Data"] = pd.to_datetime(df["Data"], dayfirst=True)

data = dict(
    type='choropleth',
    locations=['Brazil'],
    locationmode='country names',
    z=[df["USD_BRL"].mean()],
    colorscale='Viridis',
    colorbar_title="Cotação média"
)

fig = go.Figure(data=[data])
fig.update_layout(
    title_text="Cotação média do dólar no Brasil",
    title_font=dict(size=20, color="darkblue"),
    geo=dict(showframe=False, showcoastlines=True, projection_type="equirectangular")
)

fig.write_image("mapa_brasil.png", scale=2)  
fig.show()  

```
